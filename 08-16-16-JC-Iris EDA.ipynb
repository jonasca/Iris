## Iris dataset - exploring patterns using Seaborn library

This notebook analyzes the classic Iris dataset, containing length and width measurements of Sepal and Petal of three Iris species: Setosa, Versicolor and Virginica. The amazing Seaborn library was used for Exploratory Data Analysis.

```python
import numpy as np
import pandas as pd
import os
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
path = os.getcwd() + '\Iris.csv'
iris = pd.read_csv(path)
```

```python
iris.head()
```

```python
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=8) # Scatter & histogram plot at once
```

```python
sns.FacetGrid(iris, hue='Species',size=8).map(plt.scatter, 'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
sns.boxplot(data=iris, x='Species',y='PetalLengthCm')
```

```python
sns.violinplot(data=iris, x='Species',y='PetalLengthCm')
```

```python
var = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

for a in var:
    sns.FacetGrid(iris, hue='Species',size=5).map(sns.kdeplot,a).add_legend()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=2, diag_kind='kde')
```

```python
from pandas.tools.plotting import andrews_curves
from pandas.tools.plotting import parallel_coordinates
from pandas.tools.plotting import radviz
```

```python
# Andrews Curve can be used to visualize 4-dimensional data like the one presented in this dataset, which may facilitate
# understanding trends with all 4 variables. However, it can get messy in cases when the curves get too close
# such as in this example. Still beautiful though.

andrews_curves(iris.drop("Id", axis=1), "Species")
```

```python
# Parallel coordinates is another multivariate visualization technique in pandas. Plots each feature on a separate column and
# draws lines connecting the features for each data sample.

parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python
# Radviz is another multivariate visualization technique in pandas, which puts each feature as a point on a 2D plane, and
# simulates having each sample attached to those points through a spring weighted by the relative value for that feature.

radviz(iris.drop("Id", axis=1), "Species")
```
